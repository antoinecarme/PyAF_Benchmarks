```python
import pandas as pd
import numpy as np
import glob



from IPython.core.display import display, HTML
```

```python
datafiles = {}
def read_all_files():
    for f in glob.glob("../reports/*.csv"):
        benchname = f.replace("../reports/PyAF_Bench_Report_", "").replace("_debrief.csv", "")
        datafiles[benchname] = pd.read_csv(f)
    
```

```python
read_all_files()
Releases = datafiles['M4_COMP_CLIMATE']['Release'].unique()
print(Releases)
print(datafiles['M4_COMP_CLIMATE'].columns)
```

```python
def compute_index(diff_series):
    pos = diff_series[diff_series > 0]
    pos = pos.apply(lambda x : 1 if(x > 0) else 0)
    neg = diff_series[diff_series < 0]
    neg = neg.apply(lambda x : 1 if(x < 0) else 0)
    result = pos.sum() - neg.sum()
    return result


def debrief_column_evolution(iPerf):
    scores = {}
    for bench in datafiles.keys():
        df = datafiles[bench]
        df_perf = df[df['Perf'] == iPerf]
        perf_data = {}
        for rel in Releases:
            perf_data[rel] = df_perf[df_perf['Release'] == rel].reset_index()
            perf_data[rel] = perf_data[rel].set_index('Quantile')
            perf_data[rel] = perf_data[rel][['Value']]
            perf_data[rel].columns = [bench + "_" + iPerf + '_' + rel]
            # print(perf_data[rel])
        perf_data_all = pd.concat([perf_data[rel] for rel in Releases] , axis = 1) 
        # display(perf_data_all)
        scores[bench] = {}
        scores[bench][Releases[0]] = 0
        for (idx, rel) in enumerate(Releases):
            if(idx > 0):
                prev_rel = Releases[idx - 1]
                rel_data = perf_data_all[bench + "_" + iPerf + '_' + rel]
                prev_rel_data = perf_data_all[bench + "_" + iPerf + '_' + prev_rel]
                diff_rel = rel_data - prev_rel_data
                scores[bench][rel] = compute_index(diff_rel)
    rows = []
    for bench in datafiles.keys():
        row = [bench]
        for rel in Releases:
            row = row + [ scores[bench][rel] ]
        rows = rows + [row]
    df = pd.DataFrame(rows)
    # display(df)
    # print(df.columns)
    df.columns = ['bench'] + [rel for rel in Releases]
    return df
    
```
