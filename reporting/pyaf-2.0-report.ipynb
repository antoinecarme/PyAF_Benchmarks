```python
import pandas as pd
import numpy as np

from IPython.display import display, HTML

import os
import glob

%matplotlib inline
import matplotlib.pyplot as plt

```

```python
# !ls ../reporting/data
```

```python

def debrief_log_file(name, old , new):
    old_df = pd.read_csv(old)
    new_df = pd.read_csv(new)
    # print(old_df.head())
    # print(new_df.head())
    merge_df = old_df.merge(new_df , how='left', left_on='Signal', right_on='Signal', suffixes = ['_old' , '_new'])
    # print(merge_df.head())
    L1_diff =  round(merge_df['L1_new'] - merge_df['L1_old'] , 3)
    L2_diff =  round(merge_df['L2_new'] - merge_df['L2_old'] , 3)
    MAPE_diff =  round(merge_df['MAPE_new'] - merge_df['MAPE_old'] , 3)

    L1_stats = [L1_diff.min(), L1_diff.max(), L1_diff.mean(), L1_diff.std()]
    L2_stats = [L2_diff.min(), L2_diff.max(), L2_diff.mean(), L2_diff.std()]
    MAPE_stats = [MAPE_diff.min(), MAPE_diff.max(), MAPE_diff.mean(), MAPE_diff.std()]
    name1 = name.replace('_debrief.csv' , '')
    error_stats = [name1] + L1_stats + L2_stats + MAPE_stats
    
    # print(error_stats)
    return error_stats
    


def plot_error_stats(diff_df):
    return
    diff_df2 = diff_df[['Signal' , 'L1_diff', 'L2_diff' , 'MAPE_diff']]; #, 'MASE_diff']]
    diff_df2 = diff_df2.set_index('Signal' , drop=True)
    print(diff_df2.info())
    print(diff_df2.describe())
    print(diff_df2.head())
    display(diff_df2)
    plt.pcolor(1.0 - diff_df2, cmap='RdYlGn')
    plt.yticks(np.arange(0.5, len(diff_df2.index), 1), diff_df2.index)
    plt.xticks(np.arange(0.5, len(diff_df2.columns), 1), diff_df2.columns)
    fig = plt.gcf()
    fig.set_size_inches(16, min(400, int(diff_df2.shape[0] * 0.2)))
    plt.show() 
    
def debrief_directory(dirname):
    import os
    dir_df = pd.DataFrame()
    old_uri = "https://raw.githubusercontent.com/antoinecarme/PyAF_Benchmarks/master/reporting/data/"
    # "https://raw.githubusercontent.com/antoinecarme/PyAF_Benchmarks/master/reporting/"
    data = []
    for filename in sorted(glob.glob(dirname + "/*.csv")):
        try:
            base_name = os.path.basename(filename)
            old = old_uri + base_name
            new = filename
            print(new, old)
            error_stats = debrief_log_file(base_name, old , new)
            data =data + [error_stats]
        except:
            print("FAILED_READING" , filename)
            # raise
        # print(df.head())
    df = pd.DataFrame(data)
    df.columns = ['name'] + ['L1_min', 'L1_max', 'L1_mean', 'L1_std'
                                ] + ['L2_min', 'L2_max', 'L2_mean', 'L2_std'
                                    ] + ['MAPE_min', 'MAPE_max', 'MAPE_mean', 'MAPE_.std']
    print(df.head(df.shape[0]))
    plot_error_stats(df)

    return df


```
